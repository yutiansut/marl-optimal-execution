```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('millisecond.csv')
data = data[450:111508][['TIME_M', 'SIZE', 'PRICE']]
data.index = np.arange(len(data))
new_time = []
for i in range(len(data)):
    new_time.append("2019-06-28 " + data['TIME_M'][i][:-3].rjust(5, '0') + ":00")
data['TIME_M'] = new_time
data.tail(10)
```

```python
group_obj = data.groupby('TIME_M')
open_price = group_obj.first()['PRICE']
close_price = group_obj.last()['PRICE']
high_price = group_obj.max()['PRICE']
low_price = group_obj.min()['PRICE']
volumn = group_obj.sum()['SIZE']
OHLC = pd.DataFrame({'SYMBOL': 'JPM', 'level_1': open_price.index, 'open': open_price, 'high': high_price, 'low': low_price, 'close': close_price, 'volume': volumn})
OHLC.set_index(['SYMBOL', 'level_1'], inplace = True)
OHLC = OHLC.round(2)
OHLC.head()
```

```python
OHLC.to_pickle("20190628_ohlc_1m.bz2", compression="bz2")
```
